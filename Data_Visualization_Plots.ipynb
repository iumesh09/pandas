## Line PLot

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
plt.plot(x,y,'go--',linewidth=4, markersize=10)
plt.title("Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
```

## Sine Form

```python
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
plt.title("Sine Wave Form")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.plot(x,y,color=(0.7, 0.5, 0.6, 1),linewidth=2,markersize=10)
```

## Square Line Plot

```python
x = np.arange(1,11)
y = x*x
plt.plot(x,y,'m^-',linewidth=1,markersize=10)
plt.title("Square Line Plot")
plt.xlabel("Number")
plt.ylabel('Square of Number')
```

# Day 2

```python
import seaborn as sns
```

```python
x = np.arange(1,11)
y1 = x*x
y2 = x*x*x
plt.plot(x,y1,'r*-',label='Square')
plt.plot(x,y2,'c^-',label='Cube')
plt.plot()

plt.title('Line -Plot')
plt.xlabel('X-Axis')
plt.ylabel("Y-Axis")
plt.legend(loc='upper left')
plt.show()
```

# Bar Chart

```python
x = np.arange(1,11)
y = x*x
plt.bar(x,y,label="X-Square Bar Graph",color=(1,0.8,0.1))
plt.xlabel("Numbers")
plt.ylabel("Squares")
plt.legend()
plt.show()
```

# Pie Chart

```python
language = ['Python','Java','C','C++','JS']
sizes = [300,150,200,225,275]
color = ['g','b','y','r','c']
explode = (0.4,0.2,0.3,0.1,0.2)
plt.pie(sizes,labels=language,colors=color,autopct='%1.2f%%',shadow=False,explode=explode)
plt.show()
```
